```python

def chat(prompt):
    content = input("prompt: ")
    print(content)
    completion = openai.ChatCompletion.create(model="gpt-3.5-turbo", messages=[
        {"role": "system", "content": prompt},
        {"role": "system", "content": content}
    ])
    return completion.choices[0].message.content
```
